<h3> Number of Cancer clinical studies over time. </h3>

```python
# Convert 'Start Date' to datetime format
df1['Start Date']=pd.to_datetime(df1['Start Date'])

# Extract year to count studies per year
studies_per_year=df1['Start Date'].dt.year.value_counts().sort_index()
print(studies_per_year)

# Plot the results
plt.figure(figsize=(10,6))

plt.plot(studies_per_year.index, studies_per_year.values, marker='o', linestyle='-')
plt.title('Number of Cancer Clinical Studies Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Studies')
plt.grid(True)
plt.show()
```

According to this plot, **the number of studies began increasing from 2000 to 2023**, as some are expected to begin from as 2024 and even 2100. Cancer studies are still on the rise.

<h3> Clinical Trial Duration </h3>
<h4> This is to calculate the duration of clinical trials and analyze the distribution of study durations. </h4>

```python
# Calculate the duration of clinical trials in a day
df1['Start Date']=pd.to_datetime(df1['Start Date'])
df1['Completion Date']=pd.to_datetime(df1['Completion Date'])
```

```python
df1['Duration (Days)']=(df1['Completion Date'] - df1['Start Date']).dt.days
```

```python
# Summary statistics of study durations
duration_stats=df1['Duration (Days)'].describe()
```

```python
# Histogram of study durations
plt.figure(figsize=(8,6))
plt.hist(df1['Duration (Days)'], bins=40, edgecolor='k')
plt.title('Distribution of Clinical Trial Durations in Cancer Studies')
plt.xlabel('Duration (Days)')
plt.ylabel('Number of Studies')
plt.grid(True)
plt.show()

# Display summary statistics
print(duration_stats)
```

<h3> Sponsorship Trends Over Time: </h3>
Analyzing how the sponsorship of cancer clinivcal trials has evolved over the years

```python
# Extract the year from Start Date and group by year and sponsor type
sponsorship_trends=df1.groupby([df1['Start Date'].dt.year, 'Funder Type']).size().unstack().fillna(0)
```

```python
# Plot a stacked bar chart to visualize sponsorship trends
sponsorship_trends.plot(kind='bar', stacked=True, figsize=(12,6))
plt.xlabel('Year')
plt.ylabel('Number of Studies')
plt.legend(title='Funder Type', loc='upper left')
plt.show()
```
